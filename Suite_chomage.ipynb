```python
import pandas as pd
data = pd.read_excel("/home/wathelet/projet_data/python_chomage/chomage_zone.xls", "txcho_ze", skiprows = 5)
data
```

```python
#En utilisant groupby() et mean()successivementcalculer le taux de chômage moyen par régions
#pour chaque trimestres.
#Dans le groupbymettre as_index=‘False’. L’assigner à data_grp.
data_grp = data.groupby('LIBREG',as_index=False).mean()
data_grp
```

```python
#Utiliser pandas.plot() pour tracer l’histogramme du taux de chômage du premier trimestre 2020.
#Les noms de régions doivent apparaitre en abscisse.

%matplotlib inline
import matplotlib.pyplot as plt
data_grp.plot.bar(y='2020-T1', x='LIBREG', title = 'taux de chômage du premier trimestre 2020', grid=True, figsize=(20 , 10))
```
